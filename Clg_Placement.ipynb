```python
placed_mean = profile[profile["status"] == "Placed"].etest_p.mean()
unplaced_mean = profile[profile["status"] == "Not Placed"].etest_p.mean()
```

```python
def impute_etest(cols):
    etest_p = cols[0]
    status = cols[1]
    
    if pd.isnull(etest_p):
        if status == "Placed":
            return placed_mean
        elif status == "Not Placed":
            return unplaced_mean
    else:
        return etest_p
    
```

```python
profile["etest_p"]= profile[["etest_p", "status"]].apply(impute_etest, axis=1)
```

## Percentage of students who got placed in both gender

```python
profile.head()
```

```python
profile.columns
```

```python
# total male and female counts  
male_female_count = profile[["gender","status"]].groupby(by= "gender").count()
male_count= male_female_count._get_value("M", "status")
female_count= male_female_count._get_value("F", "status")

```

```python
# pass_count by gender
pass_count = profile.pivot_table(index= "gender", columns= "status" ,aggfunc= "count")
pass_count

```

```python
Male_placed = round((100/ male_count) *100 ,2 )
Female_placed = round((48/ female_count) *100 , 2)
print("Male placed: {}%".format(Male_placed))
print("Female placed: {}%".format(Female_placed))
```

## 4. Percentage of students placed accd. to hsc_s

```python
hsc_stream= profile.groupby("hsc_s").status.count()
Arts= hsc_stream[0]
Commerce= hsc_stream[1]
Science = hsc_stream[2]
```

```python
profile.pivot_table(index= "hsc_s", columns= "status", aggfunc= "count" )
```

```python
Arts_placed_percent= round((6 / Arts) *100, 2)
Commerce_placed_percent= round((79 / Commerce) *100, 2)
Science_placed_percent = round((63 / Science) *100, 2)
print("Arts placed: {}% ". format(Arts_placed_percent))
print("Commerce placed: {}% ". format(Commerce_placed_percent))
print("Science placed: {}% ". format(Science_placed_percent))
```
